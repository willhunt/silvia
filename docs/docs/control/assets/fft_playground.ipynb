# Test FFT on known input

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dummy data to check FFT
t_start = 0
t_end = 1000
f_sample = 1  # Sample frequency [Hz]
f_1 = 0.01  # Hz
f_2 = 0.02  # Hz
t = np.arange(t_start, t_end, 1/f_sample)
T = np.sin(2 * np.pi * f_1 * t) * 20 + np.sin(2 * np.pi * f_2 * t) * 15
plt.plot(t, T)
plt.xlabel('time [s]')
plt.ylabel('temperature [C]')
```

```python
n = t.shape[-1]
fft_y = np.fft.rfft(T)
fft_x = np.fft.rfftfreq(n)

plt.plot(fft_x, fft_y.real, fft_x, fft_y.imag)
```

```python
plt.plot(fft_x * f_sample, fft_y.real)
plt.xlim(0, f_2 * 2)
plt.xlabel('frequency [Hz]')
plt.plot([f_1, f_1], plt.ylim(), '--')
plt.plot([f_2, f_2], plt.ylim(), '--')
```
